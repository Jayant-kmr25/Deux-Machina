```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

Getting data

```python
url = 'https://raw.githubusercontent.com/Chef-strash/Deux-Machina/refs/heads/main/Week-2/Dataset/extended_salary_data.csv'
df = pd.read_csv(url)
# print(df.head())
years = df['YearsExperience']
X = np.array(years).reshape(-1, 1)
salary = df['Salary']
# print(salary)
```

Training mode


```python

model = LinearRegression()
model.fit(X, salary)
print("Slope: ", model.coef_)
print("Intercept: ", model.intercept_)
mae = mean_absolute_error(X, salary)
mse = mean_squared_error(X, salary)
print("MAE: ", mae)
print("MSE: ", mse)
```

Graph


```python
import matplotlib.pyplot as plt
plt.plot(years, salary)
plt.xlabel("Experience{Years}")

plt.ylabel("Salary")
plt.show()
```
